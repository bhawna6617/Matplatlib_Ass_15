```python
# defination of matplotlib :-Matplotlib is a popular Python library used for creating 2D and basic 3D plots and visualizations. It provides a wide range of tools for constructing graphs, charts, and other visual representations of data. Matplotlib is widely used in data analysis, scientific computing, and various fields such as machine learning and data science to visualize data and gain insights from it.
# use :-The Pyplot module of Matplotlib is a collection of functions that provide a simple and easy-to-use interface for creating various types of plots and customizing them. Here are five common types of plots that can be created using the Pyplot module of Matplotlib:
# five plot :-
# 1. Line Plot: Line plots are used to visualize data points connected by lines, often used for time series data or continuous data trends.

# 2. Scatter Plot: Scatter plots are used to display individual data points as dots on a 2D plane, allowing you to explore the relationships between two variables.

# 2. Bar Plot: Bar plots are used to represent categorical data with rectangular bars. They are commonly used to compare the values of different categories.

# 3. Histogram: Histograms are used to visualize the distribution of a single variable by dividing it into bins and displaying the frequency or density of data points within each bin.

# 3. Pie Chart: Pie charts represent data as a circular chart divided into sectors, where each sector represents a proportion of the whole. They are often used to show the composition of a dataset
```

```python
# scatter plot:-A scatter plot is a type of data visualization that displays individual data points as dots on a two-dimensional plane. It is particularly useful for visualizing the relationships and patterns between two continuous variables.

import numpy as np
import matplotlib.pyplot as plt


np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))


plt.scatter(x, y)


plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")


plt.show()
```

```python
# subplot:-The subplot() function in Matplotlib is used to create multiple subplots or axes within a single figure. It allows you to arrange and display multiple plots in a grid or other custom layouts within the same figure.
import numpy as np
import matplotlib.pyplot as plt


x = np.array([0, 1, 2, 3, 4, 5])


y1 = np.array([0, 100, 200, 300, 400, 500])


y2 = np.array([50, 20, 40, 20, 60, 70])

y3 = np.array([10, 20, 30, 40, 50, 60])


y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(12, 8))


plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line Plot 2')


plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line Plot 3')


plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line Plot 4')





plt.show()
```

```python
# barplot:-A bar plot, also known as a bar chart or bar graph, is a data visualization technique used to represent categorical data with rectangular bars.
# uses:-
# 1. Comparison: They make it easy to compare the values of different categories or groups at a glance.
# 2. Distribution: Bar plots can show the distribution of data within categories.
# 3. Ranking: You can use bar plots to rank categories based on their values.
# 4. Part-to-Whole Relationships: They are suitable for displaying part-to-whole relationships, such as the composition of a whole in terms of its components.

import numpy as np
import matplotlib.pyplot as plt


company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='r')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Company Profits (Vertical Bar Plot)')
plt.show()


plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='b')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.title('Company Profits (Horizontal Bar Plot)')

plt.show()

```

```python
# boxplot:-A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution and statistical summary of a dataset.
# why use?
# 1. Visualize Data Spread: Box plots show the spread of data by displaying the interquartile range (IQR), which represents the middle 50% of the data. This helps you understand the variability in your data.

# 2. Identify Central Tendency: The line inside the box represents the median, which is the central value of the data. It gives you an idea of the dataset's central tendency.

# 3. Detect Outliers: Outliers, which are data points significantly different from the rest, can be identified as individual points outside the "whiskers" of the box plot.

# 4. Compare Distributions: Box plots allow you to compare the distributions of multiple datasets side by side, making it easy to spot differences in medians, spreads, and outlier patterns.


import numpy as np
import matplotlib.pyplot as plt


box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)


plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'], notch=True, sym='gD')
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot of Two Datasets')
plt.grid(True)
plt.show()

```
